# Traning Log Visualization

```python
import matplotlib.pyplot as plt
```

```python
train_epochs = []
train_loss = []
val_epochs = []
val_loss = []

with open('log/log.txt', 'r') as file:
    for line in file:
        parts = line.split()
        step = int(parts[0])
        mode = parts[1]
        loss = float(parts[2])
        
        if mode == 'train':
            train_epochs.append(step)
            train_loss.append(loss)
        elif mode == 'val':
            val_epochs.append(step)
            val_loss.append(loss)


plt.figure(figsize=(10, 5))
plt.plot(train_epochs, train_loss, label='Training Loss', color='blue')
plt.plot(val_epochs, val_loss, label='Validation Loss', color='red')

plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training and Validation Loss over Steps')
plt.legend()
plt.grid(True)
plt.show()
```
