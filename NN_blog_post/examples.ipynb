```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample input data
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])

# Weights (randomly initialized for demonstration)
W = np.array([0.5, 0.2])
print(W)
# Calculate output
output = np.dot(X, W)
```

```python
# Visualize input data
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('Input Data')
plt.show()

# Visualize output
plt.scatter(range(len(output)), output)
plt.xlabel('Sample')
plt.ylabel('Output')
plt.title('Output')
plt.show()
```
